```python
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = [1, 2, 3]
b = [4, 5, 6]
```

```python
a, b = np.meshgrid(a, b)
```

```python
print(a)
print(b)
```

```python
fig = plt.figure()
axes = fig.gca(projection = '3d')
axes.plot_surface(a, b, a+b, cmap = "rainbow")
plt.show()
```

```python
fig = plt.figure()
axes = fig.gca(projection = '3d')
axes.plot_surface(a, b, a**2+b**2, cmap = "rainbow")
plt.show()
```
