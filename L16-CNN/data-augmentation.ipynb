```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import keras
```

```python
from keras.preprocessing import image
```

```python
img = image.load_img('../images/cat1.jpg', target_size=(100, 100))
```

```python
x = image.img_to_array(img)
```

```python
print(x.shape)
```

```python
x /= 255.0
```

```python
plt.imshow(img);
```

```python
x = x.reshape((1, 100, 100, 3))
```

```python
print(x.shape)
```

#### generating image

```python
datagen = image.ImageDataGenerator(
    rotation_range=180,
    width_shift_range=0.1,
    height_shift_range=0.1,
    shear_range=0.1,
    zoom_range=0.1,
    horizontal_flip=True,
    fill_mode='nearest'
)
```

```python
i = 0
for batch in datagen.flow(x,  batch_size=1):
    plt.figure()
    img_plot = plt.imshow(image.img_to_array(batch[0]))
    i += 1
    
    if i == 10:
        break
    plt.show()
    
```
