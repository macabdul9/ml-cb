```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import keras
```

```python
path = '/home/macab/development/datasets/corn/'
```

```python
from keras.preprocessing import image
```

```python
img = image.load_img(path + '/Healthy/7f350cf2-c946-4546-a5e0-1ae22389530c___R.S_HL 8146 copy 2_90degFlipLR.jpg', target_size=(100, 100))
```

```python
x = image.img_to_array(img)
```

```python
print(x.shape)
```

```python
x /= 255.0
```

```python
plt.imshow(img);
```

```python
x = x.reshape((1, 100, 100, 3))
```

```python
print(x.shape)
```

#### generating image

```python
datagen = image.ImageDataGenerator(
    rotation_range=180,
    width_shift_range=0.1,
    height_shift_range=0.1,
    shear_range=0.1,
    zoom_range=0.1,
    horizontal_flip=True,
    fill_mode='nearest'
)
```

```python
augmented_image = []
```

```python
i = 0
for batch in datagen.flow(x,  batch_size=1):
    augmented_image.append(image.img_to_array(batch[0]))
    i += 1
    
    if i == 10:
        break
    
    
```

```python
fig=plt.figure(figsize=(15, 15))
columns = 5
rows = 5
k = 0
for i in range(1, columns*rows +1):
    fig.add_subplot(rows, columns, i)
    plt.imshow(augmented_image[k])
    plt.title(str(k))
    k += 1
    if k > len(augmented_image) - 1:
        break
plt.show()
```
