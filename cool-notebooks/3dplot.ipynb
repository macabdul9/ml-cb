```python
from mpl_toolkits import mplot3d

import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_moons
```

```python
data, label = make_moons(n_samples=400, random_state=2, noise=0.08 )
```

```python
plt.scatter(data[:, 0], data[:, 1],s=50, c=label)
```

```python
add_feature = (10*data[:,  0]**4 + 10*data[:, 1]**4).reshape(-1, 1)
```

```python
add_feature.shape
```

```python
data = np.hstack((data, add_feature))
```

```python
fig = plt.figure(figsize=(10,  10))
ax = plt.axes(projection="3d")
plt.scatter(data[:,  0], data[:, 1], data[:, 2],c=label, marker='*')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
