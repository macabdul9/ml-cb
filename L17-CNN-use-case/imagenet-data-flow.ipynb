```python
import numpy 
from keras.preprocessing import image
import keras
import matplotlib.pyplot as plt
```

```python
path_to_imgnet = '/home/macab/development/datasets/tiny-imagenet-200/'
```

```python
train_img_gen = image.ImageDataGenerator(
    rescale=1/255.0,
    rotation_range=180,
    width_shift_range=0.3,
    height_shift_range=0.3,
    shear_range=0.2,
    zoom_range=0.3,
)


test_img_gen = image.ImageDataGenerator(
    rescale=1/255.0
)
```

#### flow from the generator

```python
train_gen = test_img_gen.flow_from_directory(path_to_imgnet + 'train', target_size=(200, 200), batch_size=64)
```
