```python
import keras
```

```python
from keras.datasets import boston_housing
```

```python
Train, Test = boston_housing.load_data()
```

```python
XT, YT = Train
print(XT.shape, YT.shape)
```

```python
Xt, Yt = Test
print(Xt.shape, Yt.shape)
```

```python
mean = XT.mean(axis = 0)
std = XT.std(axis = 0)
# data normalization
XT = (XT - mean)/std
Xt = (Xt - mean)/std
```

```python
print(mean)
```

```python
# build neural network model
from keras.models import Sequential
from keras.layers import Dense
from keras import backend as K
```

```python
def r2_score(y_true, y_pred):
    num =  K.sum(K.square(y_true - y_pred))
    denom = K.sum(K.square(y_true - K.mean(y_true)))
    return (1 - num/denom)
def get_model():
    model = Sequential()
    model.add(Dense(64, activation ='relu', input_shape = (13, )))
    model.add(Dense(32, activation ='relu'))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss= 'mse', metrics = [r2_score])
    return model
    
```

```python
model = get_model()
```

```python
model.summary()
```

```python
hist = model.fit(XT, YT, validation_data = (Xt, Yt), epochs=30)
```

```python
val_loss = hist.history['val_loss']
```

```python
train_loss = hist.history['loss']
```

```python
import matplotlib.pyplot as plt
plt.plot(val_loss)
plt.plot(train_loss)
plt.show()
```

```python
r2_score = hist.history['r2_score']
val_r2_score = hist.history['val_r2_score']
plt.plot(r2_score)
plt.plot(val_r2_score)
plt.show()
```
